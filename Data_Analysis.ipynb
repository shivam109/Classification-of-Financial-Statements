```python
import os
import pandas as pd

# Function to convert HTML table to DataFrame
def html_table_to_dataframe(html_file_path):
    # Read the HTML file
    with open(html_file_path, 'r', encoding='utf-8') as file:
        html_content = file.read()
    
    # Use pandas to read the HTML content
    dfs = pd.read_html(html_content)
    
    # Assuming you want to convert the first table found in the HTML
    if dfs:
        df = dfs[0]
        return df
    else:
        raise ValueError("No tables found in the HTML file")

# Function to clean and preprocess the DataFrame
def clean_balance_sheet_dataframe(df):
    # Remove empty columns
    df = df.dropna(axis=1, how='all')

    # Remove empty rows
    df = df.dropna(axis=0, how='all')

    # Clean up the data
    def clean_text(x):
        if isinstance(x, str):
            return x.replace('\n', '').replace('\xa0', '').replace(',', '')
        return x

    df = df.applymap(clean_text).fillna(0)

    # Identify numeric columns by inspecting the DataFrame
    print("Columns before conversion:", df.columns)
    # Ensure all columns are strings for the check
    numeric_columns = [col for col in df.columns if isinstance(col, str) and 'As at' in col]
    print("Numeric columns identified for conversion:", numeric_columns)

    # Convert numeric columns to float
    df[numeric_columns] = df[numeric_columns].apply(pd.to_numeric, errors='coerce')

    return df

# Function to process all HTML files in a directory
def process_all_html_files(directory):
    all_dataframes = []
    for subdir, dirs, files in os.walk(directory):
        for file in files:
            if file.endswith('.html'):
                file_path = os.path.join(subdir, file)
                try:
                    df = html_table_to_dataframe(file_path)
                    cleaned_df = clean_balance_sheet_dataframe(df)
                    all_dataframes.append((file_path, cleaned_df))
                    print(f"Processed {file_path}")
                except Exception as e:
                    print(f"Error processing {file_path}: {e}")
    return all_dataframes

# Directory containing all subfolders with HTML files
data_directory = 'data'

# Process all HTML files
all_dataframes = process_all_html_files(data_directory)

# Optionally, save each cleaned DataFrame to a CSV file
for file_path, df in all_dataframes:
    # Create a corresponding CSV file name
    csv_file_name = os.path.splitext(file_path)[0] + '.csv'
    # Ensure the directory exists
    os.makedirs(os.path.dirname(csv_file_name), exist_ok=True)
    # Save the DataFrame to CSV
    df.to_csv(csv_file_name, index=False)
    print(f"Saved {csv_file_name}")

# Print some information about the processed DataFrames
print(f"Processed {len(all_dataframes)} HTML files.")
```

# Model Building and Evaluation

```python
#importing all the required libraries
import os
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import classification_report
import joblib

# Combine CSV files and label them
# define a function that takes a list of directories as input and returns two lists: data (containing text data from CSV files) and labels (corresponding labels).
def load_data(directories):
    #Initializes an empty list to store text data and labels to store labels
    data = []
    labels = []
    #Loops through each directory, with label being the index and directory the directory name.
    for label, directory in enumerate(directories):
        for file in os.listdir(directory):
            if file.endswith('.csv'):
                file_path = os.path.join(directory, file)
                df = pd.read_csv(file_path)
                #Converts the DataFrame to a string
                text_data = df.to_string()
                data.append(text_data)
                labels.append(label)
    #Returns the data and labels lists.
    return data, labels

# Directories with data
directories = ["data/Income Statement", "data/Balance Sheets", "data/Cash Flow", "data/Notes", "data/Others"]

# Corresponding category names '
categories = ["Income Statement", "Balance Sheets", "Cash Flow", "Notes", "Others"]

data, labels = load_data(directories)

# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(data, labels, test_size=0.2, random_state=42)

# Feature extraction

#Initializes the TF-IDF vectorizer.
vectorizer = TfidfVectorizer()

#Fits the vectorizer to the training data and transforms the training data into TF-IDF features.
X_train_tfidf = vectorizer.fit_transform(X_train)

#Transforms the testing data into TF-IDF features using the already fitted vectorizer.
X_test_tfidf = vectorizer.transform(X_test)

# Dictionary to store models and their names as keys
models = {
    "Logistic Regression": LogisticRegression(max_iter=200),
    "SVM": SVC(),
    "Random Forest": RandomForestClassifier(n_estimators=100, random_state=42),
    "Gradient Boosting": GradientBoostingClassifier(random_state=42),
    "Neural Network": MLPClassifier(max_iter=300, random_state=42)
}

```

```python
# Train and evaluate each model
results = {}
for name, model in models.items():
    print(f"Training {name}...")
    model.fit(X_train_tfidf, y_train)
    y_pred = model.predict(X_test_tfidf)
    report = classification_report(y_test, y_pred, target_names=categories, output_dict=True)
    results[name] = report
    print(f"Classification Report for {name}:")
    print(classification_report(y_test, y_pred, target_names=categories))

```

 Out of all the models , SVM outperforms everyone. So, we will use SVM for our classification

```python
# Save the best model (SVM) and the vectorizer
joblib.dump(models['SVM'], 'svm_model.joblib')
joblib.dump(vectorizer, 'tfidf_vectorizer.joblib')
```

```python
import joblib
import pandas as pd

# Load the saved SVM model and vectorizer
svm_model = joblib.load('svm_model.joblib')
vectorizer = joblib.load('tfidf_vectorizer.joblib')

# Function to classify a new CSV file
def classify_new_file(file_path, model, vectorizer, categories):
    df = pd.read_csv(file_path)
    text_data = df.to_string()
    text_tfidf = vectorizer.transform([text_data])
    prediction = model.predict(text_tfidf)
    return categories[prediction[0]]

# Classify a new file
new_file_path = "data/18630219_table_52.csv"
predicted_category_name = classify_new_file(new_file_path, svm_model, vectorizer, categories)
print(f"The new file belongs to the category: {predicted_category_name}")
```
